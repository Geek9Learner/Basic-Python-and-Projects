# Range function and it's uses:

# It returns a sequence of numbers and is immutable (whose value is fixed).
# The range function takes one or at most three arguments, namely the start and a stop value along with a step size.
# Syntax:
#    range(stop_value) : This by default considers the starting point as zero.
#    range(start_value, stop_value) : This generates the sequence based on the start and stop value.
#    range(start_value, stop_value, step_size): It generates the sequence by incrementing the start value using the step size until it    reaches the stop value.

```python
# Question :Bond-007 , you will be given a list of numbers and we need to find 0,0,7. if it is there then print true otherwise false
def james_bond(test_list):
    bond_list=[0,0,7,'T']  # we will pop first element every time when we found pattern element.. so at last only one element will be there
    for item in test_list:
        if(item==bond_list[0]):
            bond_list.pop(0)
        else:
            pass
    return len(bond_list)==1

if __name__=="__main__":
    print(james_bond([1,2,4,0,0,7,4,0]))
    print(james_bond([1,2,3,5,0,2,5,7,0]))
    print(james_bond([1,0,2,4,0,5,7]))
```

```python
#write prime number in a range
def check_prime(num):
    prime=[2];x=3
    if(num<=2):
        return prime
    while(x<=num):
        for y in prime:
            if(x%y==0):
                break
        else:
            prime.append(x)
        x=x+2
    return prime

if __name__=="__main__":
    num=int(input('Enter the Number: '))
    prime_list=check_prime(num)
    print(prime_list)
```

# Lambda function: A anonymous block of code that used to write an annonymous function.
Helps to calulate in higher order function

```python
# Write a code to print square sum of a list element
def squareSum(numlist):
    temp=list(map(lambda x:x**2,numlist))
    print(temp)
    return sum(temp)

if __name__=="__main__":
    n=int(input("Enter the range: "))
    numList=[num for num in range(n+1)]
    square_sum_to_n=squareSum(numList)
    print(square_sum_to_n)
```

```python
# write a code to take the square sum of only even number in a given range
def check_even_sq(numlist):
    temp_list=list(filter(lambda num:num%2==0,numlist))  # filtering the even number only
    #print(temp_list)
    square_temp=list(map(lambda x:x**2,temp_list)) # mapping the square of even number
    #print(square_temp)
    return sum(square_temp)

#main Function
if __name__=="__main__":
    n=int(input("Enter the range: "))
    mylist=[x for x in range(n+1)]
    #print(mylist)
    even_square_sum=check_even_sq(mylist)
    print('Square sum of even num upto {} is: {}'.format(n,even_square_sum))
```

```python
# Write a program to print unique element without using set and dictionary
def remove_duplicate(numlist):
    result=[]
    for item in numlist:
        if item in result:
            continue
        else:
            result.append(item)
            
    print('After removing the duplicates: ',result)

#main function    
if __name__=="__main__":
    numlist=[]
    n=int(input("Enter the length of list:"))
    print('Enter the list value: ')
    for i in range(0,n):
        a=int(input())
        numlist.append(a)
    print('List element before duplicate removal:',numlist)
    remove_duplicate(numlist)
```

# Polymorphism : the word polymorphism means having many forms, So in programing language it means the same function name can be used for different signature in classes.

```python
#Example of Polymorphism -I
class India():
    def __init__(self):
        print('Inside India')
    def capital(self):
        print('Capital of india is New Delhi')
    def language(self):
        print('People speaks different language mostly hindi')
    def countryType(self):
        print('India is a developing country')
        
# different class below with same function name but different signature
class Russia():
    def __init__(self):
        print('Inside Russia')
    def capital(self):
        print('Capital of Russia is Mascow')
    def language(self):
        print('People speaks mostly Russian which beolngs to slyvic family')
    def countryType(self):
        print("It's a developed country")
        
#making Object of classes
obj_india = India()
obj_russia = Russia()
for country in (obj_india,obj_russia):
    print(type(country))
    country.capital()
    country.language()
    country.countryType()
    
#in the above example we can see that all methods with same name but different signature in both class and no conflict. this is polymorphism
```

```python
# Example of polymorphism - II
class dog():
    def __init__(self,name):
        self.name=name
    def speak(self):
        return self.name + ' says woof!!'

class cat():
    def __init__(self,name):
        self.name=name
    def speak(self):
        return self.name + ' says Meow!!'

d=dog('Mukhiya')
c=cat('Billi')
# both the speak method have different signature
print(d.speak())
print(c.speak())
```

```python
# Polymorphism with inheritance example III
class Bird:
    def intro(self):
        print("There are many types of birds.")

    def flight(self):
        print("Most of the birds can fly but some cannot.")

class sparrow(Bird):
    def flight(self):
        print("Sparrows can fly.")

class ostrich(Bird):
    def flight(self):
        print("Ostriches cannot fly.")

# from above bird class we will inherit intro() method

obj_bird = Bird()
obj_spr = sparrow()
obj_ost = ostrich()

obj_bird.intro()
obj_bird.flight()

obj_spr.intro()
obj_spr.flight()

obj_ost.intro()
obj_ost.flight()
```

```python
# Generator example - II Factorial of numbers
def factorial(n):
    for i in range(n):
        num=1
        for j in range(1,i+1):
            num=num*j
        yield num
```

# Regular expression symbols

# Symbol   Meaning            Pattern            possible match
# \d ->  A digit        -->   file_\d\d            file_25
# \w ->  A Alphanumeric -->   \w-\w\w\w            A-b_1
# \s ->  A whitespace   -->   a\sb\s\c             a b c
# \D ->  A Non-digit    -->   \D\D\D               ABC
# \W ->  Non Alphanumeric-->  \W\W\W\W\W           *+-=
# \S ->  Non whitespace -->   \S\S\S\S             Yoyo

# Quantifiers in Regular Expression

# Characters  Description               Example Pattern    Example Match
#  +          one or more times          file \w-\w+        file A-b1_1
# {3}         Exactly 3 times            \D{3}              ABC
# {2,4}       Exactly 2 to 4 times       \d{2,4}            123
# {3,}        3 or more times            \w{3,}             anycharacter
# *           zero or more times          A*B*C*             AAACC
# ?           Once or none              Plurals?           Plural

timeit runs your snippet of code millions of times (default value is 1000000) so that you get the statistically most relevant
measurement of code execution time!

timeit is pretty simple to use and has a command-line interface as well as a callable one.

```python
import timeit
help(timeit.timeit)
```

The module function timeit.timeit(stmt, setup, timer, number) accepts four arguments: 

stmt which is the statement you want to measure; it defaults to pass.
setup which is the code that you run before running the stmt; it defaults to pass. 
We generally use this to import the required modules for our code.
timer which is a timeit.Timer object; it usually has a sensible default value so you dont have to worry about it.
number which is the number of executions youd like to run the stmt.

# Example of using timeit library.

```python
def func_one(num):
    return [i for i in range(num)]
```

# Web Scrapping

Web scraping is an automated method used to extract large amounts of data from websites. The data on the websites are unstructured. Web scraping helps collect these unstructured data and store it in a structured form. There are different ways to scrape websites such as online Services, APIs or writing your own code.

# How Do You Scrape Data From A Website?

When you run the code for web scraping, a request is sent to the URL that you have mentioned. As a response to the request, the server sends the data and allows you to read the HTML or XML page. The code then, parses the HTML or XML page, finds the data and extracts it.

# Libraries used for Web Scraping.

Selenium:  Selenium is a web testing library. It is used to automate browser activities.

BeautifulSoup: Beautiful Soup is a Python package for parsing HTML and XML documents. It creates parse trees that is helpful to extract the data easily.

Pandas: Pandas is a library used for data manipulation and analysis. It is used to extract the data and store it in the desired format.

```python
import requests
import bs4
result=requests.get("http://www.example.com")
print(type(result))
#print(result.text)
soup=bs4.BeautifulSoup(result.text,"lxml")    # used for taking data into some valid format... that will help us to analyze specific data
print(soup)
```

# Some tag Syntax and their result

    Syntax                       Match Result
    
soup.select('div') --->               All the element with div tag

soup.select('#some_id')  -->        Elements containing id='some_id'

soup.select('.some_class') -->      Elements containing class='some_class'

soup.select('div span')     -->     Any element named span within div element

soup.select('div > span')  -->      Any elements named span directly within a div element, with nothing in between

```python
import requests
import bs4
result=requests.get("https://en.wikipedia.org/wiki/A._P._J._Abdul_Kalam")
soup=bs4.BeautifulSoup(result.text,"lxml")
print(soup.select('.toctext'),'\n')
for item in soup.select('.toctext'):  #toctext is a css class on above website
    print(item.text)
```

# Project Book Data Scrapping From a web

Get all the 2 star rating books from website https://books.toscrape.com/

```python
import requests
import bs4
home_URL="https://books.toscrape.com/"
page_count=1
while(page_count<=10):
    if(page_count==1):
        current_URL=home_URL
    else:
        current_URL='https://books.toscrape.com/{}{}{}{}'.format('catalogue/','page-',page_count,'.html')
    print(current_URL)
    result=requests.get(current_URL)
    soup=bs4.BeautifulSoup(result.text,"lxml")
    for book_title in soup.select('.product_pod'):  # type(book_title) --> bs4 element tag so we can use select method for book_title also
        #print(type(book_title))
        if(book_title.select('.star-rating.Two')):
            print(book_title.select('a')[1]['title'])
    page_count=page_count+1
    
# Note:: book_title.select('a')[1]['title']  --selecting the data inside html tag 'a' at index 1(list behaving nature) with value 'title'
```

# Reading the pdf file with PyPDF library

```python
import PyPDF2
pdf_text=[]
f=open('film_shooted_in_village.pdf','rb')
pdf_reader=PyPDF2.PdfFileReader(f)
for num in range(pdf_reader.numPages):
    page=pdf_reader.getPage(num)
    pdf_text.append(page.extractText(page))
```
